```python
radius = 50 # m
sq = sssq[["Bewertung", "Geo Point"]]
q_rating = []

for i, s in ssbs.iterrows():
    # print(sssq["Geo Point"][0])
    tmp = sq
    # True if Strassenquerung within range (radius)
    tmp["in_range"] = sq["Geo Point"].apply(lambda x: distance(x, s["Geo Point"]).m < radius)
    tmp = tmp[tmp["in_range"] == True]
    q_rating.append(tmp["Bewertung"].sum())
```

```python

npa = np.array(q_rating)
ssbs["q_rating"] = npa.tolist()
ssbs.to_csv(r'datasets/ssbs_sq.csv', index = False)
```

```python
radius = 60 # m

def getWRating(x, s):
    # extract points from string
    points_wrong = re.findall(r'[\d]+[.][\d]+ [\d]+[.][\d]+', str(x))
    points = []
    
    for point in points_wrong:
        point = point.split()
        point = point[1] + " " + point[0]
        points.append(point)
    s = Point(s)
    
    
    # check if street is somewhat near
    
    isNear = False
    for i in range(0, len(points), 5):
        if distance(Point(points[i]), s).m < 500:
            isNear = True
            break
    # if not, don't check further
    if not isNear:
        return 0.0
    
    dist = 100
    # start from first point possible in range
    for i in range(0, len(points)):
        dist = min(dist, distance(Point(points[i]), s).m)
        if dist < radius:
            return True
        
    return False
    

fw = ssfw[["danger", "geometry"]]

dict = {'g': 0.0,'y': 0.1, 'orange' : 0.3, 'r' : 1.0, 'cyan' : 0.0}
fw.danger = [dict[item] for item in fw.danger]
w_rating = []

for i, s in ssbs.iterrows():
    # print(sssq["Geo Point"][0])
    tmp = fw
    # True if Fussweg within range (radius)
    tmp["in_range"] = fw["geometry"].apply(lambda x: getWRating(x, s["Geo Point"]) == True)
    tmp = tmp[tmp["in_range"] == True]
    w_rating.append(tmp["danger"].sum())
    print(i, w_rating[i])

    
```

```python
npa = np.array(w_rating)
ssbs["w_rating"] = npa.tolist()
ssbs.to_csv(r'datasets/ssbs_sq_w.csv', index = False)
```
